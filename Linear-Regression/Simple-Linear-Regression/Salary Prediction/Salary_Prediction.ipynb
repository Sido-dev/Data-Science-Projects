## Import all the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

## Get the dataset

```python
df = pd.read_csv(r"Salary-dataset.csv")
df
```

## Understand Data

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.boxplot(df["Salary"])
plt.plot()
```

```python
sns.boxplot(df["YearsExperience"])
plt.plot()
```

## Creating a scatter plot to visualize the relationship between salary and YearsExperience

```python
sns.scatterplot(x=df["YearsExperience"],y=df["Salary"])
```

## Model Building 

```python
x=df[["YearsExperience"]]
y=df["Salary"]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.2,random_state=42)
xtrain,xtest,ytrain,ytest
```

## Creating a Simple Linear Regression model object

```python
slr = LinearRegression()
```

```python
slr.fit(xtrain,ytrain)
```

```python
slr.coef_
```

```python
slr.intercept_
```

## Get the prediction with xtest

```python
ypred = slr.predict(xtest)
ypred
```

```python
mae = mean_absolute_error(ytest,ypred)
mae
```

```python
mse = mean_squared_error(ytest,ypred)
mse
```

```python
rmse = mean_squared_error(ytest,ypred) ** 0.5
rmse
```

```python
r2 = r2_score(ytest,ypred)
r2
```

```python
accuracy = round(r2 * 100,2)
print("Accuracy of the model :",accuracy)
```

# Conclusion :
## The Simple Linear Regression model was successfully built to predict salary using years of experience columns
## The model achieved an R² score of 95.31%, indicating a strong positive linear relationship.
## Salary is directly proportional to years of experience, meaning as experience increases, salary also tends to increase consistently.
